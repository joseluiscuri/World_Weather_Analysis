```python
#Import dependencies
import pandas as pd
import gmaps
import gmaps.datasets
import requests
from config import g_key

# Configure gmaps to use your Google API key.

gmaps.configure(api_key=g_key)
```

```python
city_data_df=pd.read_csv("Challenge/WeatherPy_vacation.csv")
city_data_df.head()
```

```python
city_data_df_us = city_data_df.loc[(city_data_df["Country"] == "US")]
city_data_df_us
```

```python
#Choosing 4 cities in the US - Bethel, Southbridge, Norfolk, Palmer
bethel = city_data_df_us.loc[(city_data_df_us["City"] == "Bethel")]
southbridge = city_data_df_us.loc[(city_data_df_us["City"] == "Southbridge")]
norfolk = city_data_df_us.loc[(city_data_df_us["City"] == "Norfolk")]
palmer = city_data_df_us.loc[(city_data_df_us["City"] == "Palmer")]

# Setting lat and lng per city.
for index, row in bethel.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    Bethel = (lat,lng)
for index, row in southbridge.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    Southbridge = (lat,lng)
for index, row in norfolk.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    Norfolk = (lat,lng)
for index, row in palmer.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    Palmer= (lat,lng)
```

```python
fig = gmaps.figure()
bethel2norfolk = gmaps.directions_layer(Bethel,Norfolk,waypoints=[Palmer,Southbridge],travel_mode="DRIVING")
fig.add_layer(bethel2norfolk)
fig
```

```python
uscities_df = pd.concat([bethel, southbridge,norfolk,palmer], join='inner') 
uscities_df
```

```python
#Template for the pop-up marker
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""

hotel_info = [info_box_template.format(**row) for index, row in uscities_df.iterrows()]

# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.

locations = uscities_df[["Lat", "Lng"]]
max_temp = uscities_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Calling the figure
fig
```
